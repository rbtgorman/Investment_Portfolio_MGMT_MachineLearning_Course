```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dates
plt.rcParams['figure.figsize'] = [20, 15]

fs1 = pd.read_stata(r"C:\Users\rdg83\OneDrive - Rutgers University\Course Investment Portfolio Management\Week 3 Stuff\QFS.dta")

#datadate is the fiscal quarter end date.

fs1.columns

fs2=fs1[fs1['curncdq']=='USD']

#keep obs with positive assets
fs3=fs2[fs2['atq']>0]


#Drop duplicated obs
fs3.drop_duplicates(subset =['gvkey' ,'datadate'], keep = 'first', inplace = True) 

fs3['Year']=fs3['datadate'].dt.year

fs3['Month']=fs3['datadate'].dt.month



""" Macro economic data"""

#interest rate

ir1=pd.read_excel(r"C:\Users\rdg83\OneDrive - Rutgers University\Course Investment Portfolio Management\Week 3 Stuff\TB.xlsx")


ir1.dropna(inplace=True)

plt.plot(ir1['Date'],ir1['DGS10'], color='green', linewidth=2)
plt.plot(ir1['Date'],ir1['DGS3MO'], color='red', linewidth=2)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Treasury Yield in Percentage', fontsize=20)
plt.title('Treasury Yield', fontsize=20)
plt.legend(['10-Year','3-Month'], fontsize=20)
plt.gca().xaxis.set_major_locator(dates.YearLocator(5, month=12, day=31))


ir1['spread']=ir1['DGS10']-ir1['DGS3MO']

ir1['zero']=0


plt.plot(ir1['Date'],ir1['spread'], color='grey', linewidth=2)
plt.plot(ir1['Date'],ir1['zero'], color='blue', linewidth=2, linestyle='dashed')
plt.xlabel('Date', fontsize=20)
plt.ylabel('Yield spread in Percentage', fontsize=20)
plt.title('Yield spread', fontsize=20)
plt.legend(['Yield spread'], fontsize=20)
plt.gca().xaxis.set_major_locator(dates.YearLocator(5, month=12, day=31))
'''



#GDP growth rate

gdp1=pd.read_excel((r"C:\Users\rdg83\OneDrive - Rutgers University\Course Investment Portfolio Management\Week 3 Stuff\GDP-QGR.xlsx"), sheet_name='real gdp growth rate')
#The advance estimate of quartelry GDP is released about one month after the quarter end
#Release schedule https://www.bea.gov/news/schedule/full

plt.plot(gdp1['Date'],gdp1['GDP_Growth'], color='green', linewidth=2)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Quarterly GDP Growth Rate in Percentage', fontsize=20)
plt.title('Quarterly GDP Growth Rate in Percentage', fontsize=20)
plt.gca().xaxis.set_major_locator(dates.YearLocator(5, month=12, day=31))





""" Merge Example"""


sample1=pd.read_stata((r"C:\Users\rdg83\OneDrive - Rutgers University\Course Investment Portfolio Management\Week 3 Stuff\QFS.dta"))


sample2=pd.read_stata((r"C:\Users\rdg83\OneDrive - Rutgers University\Course Investment Portfolio Management\Week 3 Stuff\penndataHW3.dta"))


sample_inner1=pd.merge(sample1, sample2,left_on=['Name'], right_on=['Name'], how='inner')
#left_on and right_on indicate the variables to link two datasets
#inner join keep the intersection of two datasets


sample_left1=pd.merge(sample1, sample2,left_on=['Name'], right_on=['Name'], how='left')
#left join keep all the information in the left datasets

sample_right1=pd.merge(sample1, sample2,left_on=['Name'], right_on=['Name'], how='right')
#right join keep all the information in the right datasets


sample_outer1=pd.merge(sample1, sample2,left_on=['Name'], right_on=['Name'], how='outer')
# outer join keep the union of two datasets




"""
Merge returns, financial statement data, and macroeconomic data
"""

"""
Quarterly financial statement information is from Form 10-Q, available in SEC edgar

Companies can dely their reporting to SEC edgar. But there are deadlines.

https://www.investor.gov/introduction-investing/investing-basics/glossary/form-10-k


To ensure investors can access the financial statement information,
we require that there are at least 90-day gap between fiscal quarter end 
and the end of the previous month.Thus, we can obtain the financial statement
information before the beginning of a month. 

Dec:fiscal quarter end
Jan: gap
Feb: gap
Mar: gap
April: month of return

"""

st7=pd.read_stata(r"C:\Users\rdg83\OneDrive - Rutgers University\Course Investment Portfolio Management\Week 3 Stuff\stock return processed.dta")


stret1=st7[['gvkey', 'ret','datadate']]

stret1.head()


stret1['lagdatadate']=stret1['datadate']+pd.Timedelta(days=-40)+pd.tseries.offsets.BusinessMonthEnd(0)
#pd.Timedelta(days=-40) find date 40 days before datadate
#find the last business day of the previous month
#pd.tseries.offsets.BusinessMonthEnd(0) returns the last business day of a month


stret1.head()


fs4=fs3[['gvkey', 'datadate','atq','ceqq','dlttq','epspiq']]

fs4.rename(columns={'datadate':'date_fs'}, inplace=True)



merge1=pd.merge(stret1, fs4, left_on=['gvkey'], right_on=['gvkey'], how='inner')
#merge based on gvkey

merge1['datediff']=merge1['lagdatadate']-merge1['date_fs']


merge2=merge1[(merge1['datediff']>pd.Timedelta(days=90))
              &(merge1['datediff']<pd.Timedelta(days=200))]


#One gvkey-date could be matched with multiple quarter end accounting info
#we just keep the most recent quarter-end financial statement information for each gvkey-date



merge2.sort_values(by=['gvkey' ,'datadate','date_fs'], inplace=True)

merge2.head()

merge2.drop_duplicates(subset =['gvkey' ,'datadate'], keep = 'last', inplace = True) 


merge2['datediff'].min()


merge2['datediff'].max()


merge2.drop(columns=('datediff'),inplace=True)#datediff format is not supported in stata

merge2.head()



#merge with treasury yield

#information is only at the Date level

#We create a list of lagdatadate for merge to reduce processing time

lagdatadate1=merge2[['lagdatadate']]

lagdatadate1.drop_duplicates(subset =['lagdatadate'], keep = 'first', inplace = True) 


lagdatadate1['link']=1

ir2=ir1.copy()

ir2['link']=1

ir2.rename(columns={'Date':'Date_ir'},inplace=True)

ir3=pd.merge(lagdatadate1,ir2,left_on=['link'], right_on=['link'],how='inner')

ir3['datediff_ir']=ir3['lagdatadate']-ir3['Date_ir']

#find the average interet rate in the previous 30 days
#treasury yield is posted everyday.
ir4=ir3[(ir3['datediff_ir']>=pd.Timedelta(days=1))
              &(ir3['datediff_ir']<pd.Timedelta(days=31))]

ir5=ir4[['lagdatadate','DGS3MO','DGS10']].groupby(['lagdatadate'], as_index=False).mean()

merge3=pd.merge(merge2, ir5, left_on=['lagdatadate'], right_on=['lagdatadate'], how='inner')


merge3.head()



#merge GDP

gdp2=gdp1.copy()

gdp2['link']=1


gdp2.rename(columns={'Date':'Date_gdp'},inplace=True)

gdp3=pd.merge(lagdatadate1,gdp2,left_on=['link'], right_on=['link'],how='inner')

gdp3['datediff_gdp']=gdp3['lagdatadate']-gdp3['Date_gdp']


##The advance estimate of quarterly GDP is released about one month after the quarter end
gdp4=gdp3[(gdp3['datediff_gdp']>=pd.Timedelta(days=25))
              &(gdp3['datediff_gdp']<pd.Timedelta(days=130))]


#only keep the most recent quarterly gdp growth rate before the lagdatadate
#If there are mutiple Date_dgp for one lagdatadate
#We sort by lagdatadate and date_gdp in ascending order
#For each lagdatadate, we keep the last Date_gdp. The last Date_gdp corresponds to the most recent quarterly gdp growth rate
gdp4.sort_values(by=['lagdatadate','Date_gdp'], inplace=True)

gdp4.drop_duplicates(subset =['lagdatadate'], keep = 'last', inplace = True) 


gdp5=gdp4[['lagdatadate','GDP_Growth']]

merge4=pd.merge(merge3, gdp5, left_on=['lagdatadate'], right_on=['lagdatadate'], how='inner')

merge4.head()


merge4.to_stata((r"C:\Users\rdg83\OneDrive - Rutgers University\Course Investment Portfolio Management\Week 3 Stuff\week 3 sample.dta"), write_index=False)

```
